```python
#Normlization of function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
#Normlized data frame (considered numeric part of data)
df_norm = norm_func(data.iloc[:,1:])
```

```python
df_norm
```

```python
#Create Dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm,method='complete'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm,method='single'))
```

```python
'''
by using hierarchical clustering (complete method) we can get 4 clusters
'''
```

# KMeans 

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
#Elbow Curve
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# selection 4 cluster from the above scree plot which is optimum number of clusters
model = KMeans(n_clusters = 4)
model.fit(df_norm)
model.labels_       #getting the number of clusters assign to each row
```

```python
md = pd.Series(model.labels_)
data['clust'] = md
data
```

```python
data.iloc[:,1:5].groupby(data.clust).mean()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data2 = data.drop('clust',axis=1)   
```

```python
array = data2.iloc[:,1:].values
```

```python
array
```

```python
stscalar = StandardScaler().fit(array)
X = stscalar.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=0.8,min_samples=4)
dbscan.fit(X)
```

```python
#Noise samples are given the label -1
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_,columns=['Cluster'])
```

```python
cl
```

```python
pd.concat([data2,cl],axis=1)
```
